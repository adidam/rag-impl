```python
from kaggle_secrets import UserSecretsClient
user_secrets = UserSecretsClient()
groq_key = user_secrets.get_secret("first_key")
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_groq import ChatGroq

chat = ChatGroq(temperature=0.3, groq_api_key=groq_key, model_name="llama3-8b-8192")

def generate_pseudo_document_llm(query: str):

    prompt = ChatPromptTemplate.from_template(
      """
      Please provide a response to the query below, in a passage.
      Respond just with the passage only.
      
      Question: {query}

      Passage:
      """
    )

    chain = prompt | chat

    context = "".join(query)

    groq_response = chain.invoke({"query": query})

    groq_response.pretty_print()

    answer = groq_response.content
    return query, answer
```

```python
from sentence_transformers import SentenceTransformer

embedder = SentenceTransformer("BAAI/LLM-Embedder")

def get_pseudo_doc_embeddings(answer: str):
     return embedder.encode(answer).tolist()
```

```python
question = "when was Rolex registered?"

answer = generate_pseudo_document_llm(question)[1]
print(f'answer {answer}')
psudo_emb = get_pseudo_doc_embeddings(answer)
psudo_emb
```
